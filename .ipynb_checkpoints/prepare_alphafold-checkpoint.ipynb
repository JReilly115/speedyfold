```python
# ============================================================================
# COMPLETE PIPELINE - Using ANARCI for proper antibody numbering
#                   - Precomputes alignments to minimize run time
# ============================================================================

import time
from tqdm import tqdm

# ----------------------------------------------------------------------------
# Set PATH to find HMMER in AlphaFold3 environment
# ----------------------------------------------------------------------------

hmmer_path = "/cluster/tufts/cowenlab/jreill05/Database_Conda_Environment/condaenv/AlphaFold3/bin"
os.environ["PATH"] = hmmer_path + ":" + os.environ.get("PATH", "")
from anarci import anarci

# ----------------------------------------------------------------------------
# PART 1: Get sequence lengths and find longest query
# ----------------------------------------------------------------------------

target_seq = list(inputs.values())[0]
target_length = len(target_seq)

binder_items = []
for name, seq in inputs.items():
    if name != 'target':
        binder_items.append((name, seq))

print(f"Target length: {target_length}")
print(f"Total binders: {len(binder_items)}")

# Find the longest binder sequence to use as reference
longest_binder = max(binder_items, key=lambda x: len(x[1]))
ref_name, ref_seq = longest_binder
ref_length = len(ref_seq)
print(f"Longest query (reference): {ref_name} (length: {ref_length})")
print("="*60)

# ----------------------------------------------------------------------------
# PART 2: Load target templates
# ----------------------------------------------------------------------------

with open(template_target_file, 'r') as f:
    target_templates = json.load(f)
print(f"Loaded {len(target_templates)} target templates")

# ----------------------------------------------------------------------------
# PART 3: ANARCI alignment functions
# ----------------------------------------------------------------------------

def collect_all_sequences(binder_items, paired_MSA_collection_path, unpaired_MSA_collection_path):
    """
    Collect ALL sequences from binders and the TWO MSA files.
    
    Args:
        binder_items: List of (name, seq) for all binders
        paired_MSA_collection_path: Path to the SINGLE paired MSA file (BINDER.a3m)
        unpaired_MSA_collection_path: Path to the SINGLE unpaired MSA file (BINDER.a3m)
    """
    all_sequences = []
    
    # Add all binder sequences
    for name, seq in binder_items:
        all_sequences.append(seq)
    
    # Read PAIRED MSA file (ONCE!)
    with open(paired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            homo_seq = lines[i + 1].strip()
            if homo_seq:
                all_sequences.append(homo_seq)
    
    # Read UNPAIRED MSA file (ONCE!)
    with open(unpaired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            homo_seq = lines[i + 1].strip()
            if homo_seq:
                all_sequences.append(homo_seq)
    
    return all_sequences

anarci_cache = {}

def get_anarci_numbering_dict(sequence, scheme='imgt'):
    """
    Arguments:
        sequence: Raw amino acid sequence (can have gaps - will be cleaned)
        scheme: Numbering scheme ('imgt', 'aho', etc.)
    
    Returns:
        Dictionary: {(pos, ins): aa, ...} or None if fails
        Example: {(1, ''): 'Q', (2, ''): 'V', (111, 'A'): 'D', ...}
    """
    # Clean sequence: remove gaps, keep only letters
    seq_clean = ''.join([c for c in sequence if c.isalpha()])
    
    # Skip sequences that are too short (not full antibodies)
    if len(seq_clean) < 50:
        return None
    
    # Check cache first - if we've seen this sequence before, return stored result
    if seq_clean in anarci_cache:
        return anarci_cache[seq_clean]
    
    # Run ANARCI (this is the slow part - happens only once per unique sequence)
    result = anarci([('seq', seq_clean)], scheme=scheme, output=False, allowed_species=[])

    if result is None:
        return None

    if isinstance(result, tuple) and len(result) >= 1:
        numbered_seqs = result[0]
      # print("Warning - multiple results per alignment found, only first used")
    else:
        numbered_seqs = result

    if not numbered_seqs or not numbered_seqs[0]:
        return None

    numbered = numbered_seqs[0][0][0]

    # Convert to dictionary for fast lookup
    pos_dict = {}
    for (pos, ins), aa in numbered:
        if aa is not None:
            pos_dict[(pos, ins)] = aa

    # Store in cache for future use
    anarci_cache[seq_clean] = pos_dict
    return pos_dict


def build_master_framework(all_sequences, scheme='imgt'):
    """
    Build a master framework containing ALL IMGT positions from all sequences.
    Uses cached ANARCI results to avoid redundant processing.
    
    Args:
        all_sequences: List of sequences (can include binders and homologs)
        scheme: Numbering scheme
    
    Returns:
        master_positions: Sorted list of all unique (pos, ins) tuples
    """
    all_positions = set()
    
    print(f"  Processing {len(all_sequences)} sequences to build master framework...")
    
    for seq in tqdm(all_sequences, desc="    Building master framework", unit="seq"):
        pos_dict = get_anarci_numbering_dict(seq, scheme)
        if pos_dict:
            for pos in pos_dict.keys():
                all_positions.add(pos)
                pos_num, pos_ins = pos
                if pos_ins != '' and pos_ins is not None:
                    all_positions.add((pos_num, ''))                     
    
    # Sort positions
    def sort_key(item):
        pos, ins = item
        ins_key = '' if ins is None or ins == '' else ins
        return (pos, ins_key)
    
    sorted_positions = sorted(all_positions, key=sort_key)
    
    # Clean: remove bare numbers that have lettered versions
    numbers_with_letters = set()
    for pos, ins in sorted_positions:
        if ins != '' and ins is not None:
            numbers_with_letters.add(pos)
    
    master_positions = []
    for pos, ins in sorted_positions:
        if ins == '' or ins is None:
            # Bare number - only keep if no lettered versions exist
            if pos not in numbers_with_letters:
                master_positions.append((pos, ins))
        else:
            # Lettered version - always keep
            master_positions.append((pos, ins))
    
    return master_positions

def align_to_master_framework(master_positions, pos_dict):
    """
    Arguments:
        master_positions: Sorted list of (pos, ins) tuples
        pos_dict: Dictionary mapping (pos, ins) to amino acid
    
    Returns:
        aligned_string: Sequence with residues at master positions, '-' elsewhere
        residue_count: Number of actual residues (for verification)
    """
    if not pos_dict:
        return None, 0
    
    # Build aligned sequence
    aligned_chars = []
    for position in master_positions:
        if position in pos_dict:
            aligned_chars.append(pos_dict[position])
        else:
            aligned_chars.append('-')
    
    aligned = ''.join(aligned_chars)
    residue_count = sum(1 for c in aligned if c.isalpha())
    
    return aligned, residue_count


def precompute_aligned_sequences(master_positions, binder_items, paired_MSA_collection_path, unpaired_MSA_collection_path, scheme='imgt'):
    """
    Precompute aligned versions of ALL sequences.
    
    Reads the SINGLE paired and SINGLE unpaired MSA files.
    """
    aligned_queries = {}
    precomputed_paired = []
    precomputed_unpaired = []
    paired_headers = []
    unpaired_headers = []
    paired_query_header = None
    unpaired_query_header = None
    
    # Track seen sequences to avoid duplicates
    seen_paired = set()
    seen_unpaired = set()
    
    # Store dictionaries for fast alignment later
    numbered_queries = {}
    
    print("  Getting numbering for binder queries...")
    for name, seq in tqdm(binder_items, desc="    Numbering binders", unit="binder"):
        pos_dict = get_anarci_numbering_dict(seq, scheme)
        if pos_dict:
            numbered_queries[seq] = pos_dict
    
    print("  Aligning binder queries to master framework...")
    for seq, pos_dict in numbered_queries.items():
        aligned, _ = align_to_master_framework(master_positions, pos_dict)
        if aligned:
            aligned_queries[seq] = aligned
    
    print("  Processing PAIRED MSA file...")
    with open(paired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    
    paired_query_header = lines[0]
    
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            header = lines[i]
            homo_seq = lines[i + 1].strip()
            
            paired_headers.append(header)
            
            raw_seq = ''.join([c for c in homo_seq if c.isalpha()])
            
            if raw_seq and raw_seq not in seen_paired:
                pos_dict = get_anarci_numbering_dict(raw_seq, scheme)
                if pos_dict:
                    aligned, _ = align_to_master_framework(master_positions, pos_dict)
                    if aligned:
                        precomputed_paired.append(aligned)
                        seen_paired.add(raw_seq)
    
    print("  Processing UNPAIRED MSA file...")
    with open(unpaired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    
    unpaired_query_header = lines[0]
    
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            header = lines[i]
            homo_seq = lines[i + 1].strip()
            
            unpaired_headers.append(header)
            
            raw_seq = ''.join([c for c in homo_seq if c.isalpha()])
            
            if raw_seq and raw_seq not in seen_unpaired:
                pos_dict = get_anarci_numbering_dict(raw_seq, scheme)
                if pos_dict:
                    aligned, _ = align_to_master_framework(master_positions, pos_dict)
                    if aligned:
                        precomputed_unpaired.append(aligned)
                        seen_unpaired.add(raw_seq)
    
    print(f"  Precomputed {len(precomputed_paired)} unique PAIRED homologs")
    print(f"  Precomputed {len(precomputed_unpaired)} unique UNPAIRED homologs")
    
    return (aligned_queries, precomputed_paired, precomputed_unpaired, 
            paired_headers, unpaired_headers, paired_query_header, unpaired_query_header)


def process_binder(binder_seq, master_positions, aligned_queries, precomputed_homologs, headers, query_header, scheme='imgt'):
    """
    Process a single binder: apply transformations and return MSA components.
    
    Arguments:
        binder_seq: Raw binder sequence
        master_positions: Master framework positions (for fallback alignment)
        aligned_queries: Dict mapping binder_seq to aligned version
        precomputed_homologs: List of precomputed aligned homologs
        headers: List of headers for homologs
        query_header: Header line for the query
        scheme: Numbering scheme
    
    Returns:
        raw_query: Raw query (for MSA line 2)
        processed_homologs: List of processed homologs ready for MSA
        msa_headers: List of headers for MSA
        msa_sequences: List of sequences for MSA
    """
    raw_query = binder_seq.upper()
    
    # Get aligned query from precomputed map
    query_aligned = aligned_queries[binder_seq]
    # if not query_aligned:
    #     # Fallback: align on the fly using dictionary method
    #     pos_dict = get_anarci_numbering_dict(binder_seq, scheme)
    #     if pos_dict:
    #         query_aligned, _ = align_to_master_framework(master_positions, pos_dict)
    #     if not query_aligned:
    #         query_aligned = raw_query
    
    # Apply dummy lowercase and substitution
    processed_homologs = add_dummy_lowercase_and_substitute(query_aligned, precomputed_homologs)
    
    # Build MSA components
    msa_headers = [query_header]
    msa_sequences = [raw_query]
    
    for i in range(min(len(headers), len(processed_homologs))):
        msa_headers.append(headers[i])
        msa_sequences.append(processed_homologs[i])
    
    return raw_query, processed_homologs, msa_headers, msa_sequences


def add_dummy_lowercase_and_substitute(aligned_query, aligned_homologs):
    """
    1. Add dummy 'x' where both query and homolog have dashes
    2. Convert uppercase to lowercase in homolog where query has a dash
    """
    processed = []
    aligned_query = aligned_query.upper()
    for homo in aligned_homologs:
        # # Ensure same length
        # if len(homo) != len(aligned_query):
        #     if len(homo) < len(aligned_query):
        #         needed = len(aligned_query) - len(homo)
        #         homo = homo + ('-' * needed)
        #     else:
        #         homo = homo[:len(aligned_query)]
        
        temp_homo = []
        homo = homo.upper()
        for i in range(len(aligned_query)):
            q_char = aligned_query[i]
            h_char = homo[i]
            
            if q_char == '-' and h_char == '-':
                # Both gaps → add 'x'    or     # pass
                pass
            elif q_char == '-' and h_char.isalpha():
                # Query gap, homolog uppercase → convert to lowercase
                temp_homo.append(h_char.lower())
            else:
                # All other cases: keep as is
                temp_homo.append(h_char)
        
        processed.append(''.join(temp_homo))
    
    return processed


def filter_msa_by_query_length(header_list, sequence_list, query_sequence):
    
    query_length = len(query_sequence)
    
    filtered_headers = [header_list[0]]
    filtered_sequences = [sequence_list[0]]
    removed = 0
    total = len(header_list) - 1  # Exclude query
    
    for i in range(1, len(header_list)):
        header = header_list[i]
        sequence = sequence_list[i]
        
        clean_sequence_chars = []
        for c in sequence:
            if not c.islower() or c == '-':
                clean_sequence_chars.append(c)
        clean_sequence = ''.join(clean_sequence_chars)
        
        if len(clean_sequence) == query_length:
            filtered_headers.append(header)
            filtered_sequences.append(sequence)
        else:
            removed += 1
    
    return filtered_headers, filtered_sequences, removed, total


def remove_gaps(sequence):
    
    result = []
    for c in sequence:
        if c.isalpha():
            result.append(c)
    return ''.join(result)


def check_indices(templates, seq_length):
    """Check that all indices in templates are within bounds."""
    all_valid = True
    for template in templates:
        if 'queryIndices' in template and template['queryIndices']:
            for i in template['queryIndices']:
                if i >= seq_length:
                    all_valid = False
                    break
        if not all_valid:
            break
            
    return all_valid


def get_safe_name(name):
    """Convert a binder name to a filesystem-safe string."""
    safe = ''
    for c in name:
        if c == '/' or c == '\\' or c == '.':
            safe += '_'
        else:
            safe += c
    return safe


def get_paired_path(safe_name):
    """Return the path to the paired MSA file for a binder."""
    return f"{MSA_collection_path}/pairedMsa/binder/{safe_name}.a3m"


def get_unpaired_path(safe_name):
    """Return the path to the unpaired MSA file for a binder."""
    return f"{MSA_collection_path}/unpairedMsa/binder/{safe_name}.a3m"

# ----------------------------------------------------------------------------
# PART 4: BUILD MASTER FRAMEWORK
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 1: Building master framework from all sequences")
print("="*60)

start_time = time.time()

all_sequences = collect_all_sequences(binder_items, pairedMsa_binder_file, unpairedMsa_binder_file)
print(f"Collected {len(all_sequences)} total sequences")

master_positions = build_master_framework(all_sequences)
print(f"Master framework has {len(master_positions)} positions")
print(f"Master framework build time: {time.time() - start_time:.1f}s")

# ----------------------------------------------------------------------------
# PART 5: PRECOMPUTE ALIGNED SEQUENCES
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 2: Precomputing aligned sequences")
print("="*60)

start_time = time.time()

(aligned_queries, precomputed_paired, precomputed_unpaired, 
 paired_headers, unpaired_headers, paired_query_header, unpaired_query_header) = precompute_aligned_sequences(
    master_positions, binder_items, pairedMsa_binder_file, unpairedMsa_binder_file)

print(f"Precomputation time: {time.time() - start_time:.1f}s")

# ----------------------------------------------------------------------------
# PART 6: Process each binder using trimmed alignments
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 3: Processing individual binders")
print("="*60)

total_start_time = time.time()

for binder_idx, (original_name, binder_seq) in enumerate(tqdm(binder_items, desc="Processing binders", unit="binder"), 1):
    binder_start_time = time.time()
    
    # Use helper function for safe name
    safe_name = get_safe_name(original_name)
    
    binder_length = len(binder_seq)
    raw_query = binder_seq.upper()
    
    print(f"\n[{binder_idx}/{len(binder_items)}] Processing {original_name} (length: {binder_length})...")
    
    # --- Create binder-specific template file with trimmed indices ---
    with open(template_binder_file, 'r') as f:
        binder_templates = json.load(f)
    
    for template in binder_templates:
        if 'queryIndices' in template:
            trimmed_indices = []
            for i in template['queryIndices']:
                if i < binder_length:
                    trimmed_indices.append(i)
            template['queryIndices'] = trimmed_indices
    
    binder_template_copy = f"{MSA_collection_path}/templates/binder/{safe_name}_templates.json"
    os.makedirs(os.path.dirname(binder_template_copy), exist_ok=True)
    
    with open(binder_template_copy, 'w') as f:
        json.dump(binder_templates, f, indent=2)
    
    # --- Process PAIRED MSA ---
    paired_start_time = time.time()
    print(f"  [PAIRED MSA] Processing...")
    
    raw_query, processed_homologs, msa_headers, msa_sequences = process_binder(
        binder_seq, master_positions, aligned_queries, precomputed_paired, 
        paired_headers, paired_query_header
    )
    
    filtered_headers, filtered_sequences, paired_removed, paired_total = filter_msa_by_query_length(
        msa_headers, msa_sequences, raw_query
    )
    
    paired_file = get_paired_path(safe_name)
    os.makedirs(os.path.dirname(paired_file), exist_ok=True)
    
    with open(paired_file, 'w') as f:
        for j in range(len(filtered_headers)):
            f.write(filtered_headers[j])
            f.write(filtered_sequences[j] + '\n')
    
    paired_passed = len(filtered_sequences) - 1
    paired_time = time.time() - paired_start_time
    print(f"  [PAIRED MSA] ✓ Passed: {paired_passed}/{paired_total} | Removed: {paired_removed} | Time: {paired_time:.1f}s")
    
    # --- Process UNPAIRED MSA ---
    unpaired_start_time = time.time()
    print(f"  [UNPAIRED MSA] Processing...")
    
    raw_query, processed_homologs, msa_headers, msa_sequences = process_binder(
        binder_seq, master_positions, aligned_queries, precomputed_unpaired, 
        unpaired_headers, unpaired_query_header
    )
    
    filtered_headers, filtered_sequences, unpaired_removed, unpaired_total = filter_msa_by_query_length(
        msa_headers, msa_sequences, raw_query
    )
    
    unpaired_file = get_unpaired_path(safe_name)
    os.makedirs(os.path.dirname(unpaired_file), exist_ok=True)
    
    with open(unpaired_file, 'w') as f:
        for j in range(len(filtered_headers)):
            f.write(filtered_headers[j])
            f.write(filtered_sequences[j] + '\n')
    
    unpaired_passed = len(filtered_sequences) - 1
    unpaired_time = time.time() - unpaired_start_time
    print(f"  [UNPAIRED MSA] ✓ Passed: {unpaired_passed}/{unpaired_total} | Removed: {unpaired_removed} | Time: {unpaired_time:.1f}s")
    
    # --- Create JSON files ---
    with open(binder_template_copy, 'r') as f:
        trimmed_binder_templates = json.load(f)

    query_no_gaps = remove_gaps(aligned_queries.get(binder_seq, binder_seq))

    json_parts = []
    json_parts.append('{\n')
    json_parts.append(f'  "name": "{original_name}",\n')
    json_parts.append('  "modelSeeds": [1, 2],\n')
    json_parts.append('  "sequences": [\n')
    json_parts.append('    {\n')
    json_parts.append('      "protein": {\n')
    json_parts.append('        "id": "TARGET",\n')
    json_parts.append(f'        "sequence": {json.dumps(target_seq)},\n')
    json_parts.append(f'        "unpairedMsaPath": {json.dumps(unpairedMsa_target_file)},\n')
    json_parts.append(f'        "pairedMsaPath": {json.dumps(pairedMsa_target_file)},\n')
    json_parts.append('        "templates": [\n')
    
    for i in range(len(target_templates)):
        template = target_templates[i]
        json_parts.append('  {\n')
        json_parts.append(f'    "mmcifPath": {json.dumps(template["mmcifPath"])},\n')
        
        query_indices_str = '['
        for j in range(len(template['queryIndices'])):
            if j > 0:
                query_indices_str += ', '
            query_indices_str += str(template['queryIndices'][j])
        query_indices_str += ']'
        json_parts.append(f'    "queryIndices": {query_indices_str},\n')
        
        template_indices_str = '['
        for j in range(len(template['templateIndices'])):
            if j > 0:
                template_indices_str += ', '
            template_indices_str += str(template['templateIndices'][j])
        template_indices_str += ']'
        json_parts.append(f'    "templateIndices": {template_indices_str}\n')
        
        if i < len(target_templates) - 1:
            json_parts.append('  },\n')
        else:
            json_parts.append('  }\n')
    
    json_parts.append(']\n')
    json_parts.append('      }\n')
    json_parts.append('    },\n')
    json_parts.append('    {\n')
    json_parts.append('      "protein": {\n')
    json_parts.append('        "id": "BINDER",\n')
    json_parts.append(f'        "sequence": {json.dumps(query_no_gaps)},\n')
    json_parts.append(f'        "unpairedMsaPath": {json.dumps(unpaired_file)},\n')
    json_parts.append(f'        "pairedMsaPath": {json.dumps(paired_file)},\n')
    json_parts.append('        "templates": [\n')
    
    for i in range(len(trimmed_binder_templates)):
        template = trimmed_binder_templates[i]
        json_parts.append('  {\n')
        json_parts.append(f'    "mmcifPath": {json.dumps(template["mmcifPath"])},\n')
        
        query_indices_str = '['
        for j in range(len(template['queryIndices'])):
            if j > 0:
                query_indices_str += ', '
            query_indices_str += str(template['queryIndices'][j])
        query_indices_str += ']'
        json_parts.append(f'    "queryIndices": {query_indices_str},\n')
        
        template_indices_str = '['
        for j in range(len(template['templateIndices'])):
            if j > 0:
                template_indices_str += ', '
            template_indices_str += str(template['templateIndices'][j])
        template_indices_str += ']'
        json_parts.append(f'    "templateIndices": {template_indices_str}\n')
        
        if i < len(trimmed_binder_templates) - 1:
            json_parts.append('  },\n')
        else:
            json_parts.append('  }\n')
    
    json_parts.append(']\n')
    json_parts.append('      }\n')
    json_parts.append('    }\n')
    json_parts.append('  ],\n')
    json_parts.append('  "dialect": "alphafold3",\n')
    json_parts.append('  "version": 4\n')
    json_parts.append('}')
    
    json_file = os.path.join(json_collection_path, f"{safe_name}.json")
    dir_path = ''
    for part in json_file.split('/')[:-1]:
        dir_path += part + '/'
    if dir_path:
        os.makedirs(dir_path, exist_ok=True)
    
    with open(json_file, 'w') as f:
        f.write(''.join(json_parts))
    
    binder_time = time.time() - binder_start_time
    print(f"  ✓ JSON created: {safe_name}.json | Binder time: {binder_time:.1f}s")
    print("-" * 60)
    
# ----------------------------------------------------------------------------
# FINAL SUMMARY
# ----------------------------------------------------------------------------

total_time = time.time() - total_start_time
print("\n" + "="*60)
print("PIPELINE COMPLETE")
print("="*60)
print(f"Total processing time: {total_time/60:.1f} minutes ({total_time:.1f} seconds)")
print(f"Output files:")
print(f"  Paired MSAs: {MSA_collection_path}/pairedMsa/binder/")
print(f"  Unpaired MSAs: {MSA_collection_path}/unpairedMsa/binder/")
print(f"  JSON files: {json_collection_path}")
print("="*60)
```

```python
# Accounting for reverse ordering of lettered sequence

# ============================================================================
# COMPLETE PIPELINE - Using ANARCI for proper antibody numbering
#                   - Precomputes alignments to minimize run time
# ============================================================================

import time
from tqdm import tqdm
from collections import defaultdict

# ----------------------------------------------------------------------------
# Set PATH to find HMMER in AlphaFold3 environment
# ----------------------------------------------------------------------------

hmmer_path = "/cluster/tufts/cowenlab/jreill05/Database_Conda_Environment/condaenv/AlphaFold3/bin"
os.environ["PATH"] = hmmer_path + ":" + os.environ.get("PATH", "")
from anarci import anarci

# ----------------------------------------------------------------------------
# PART 1: Get sequence lengths and find longest query
# ----------------------------------------------------------------------------

target_seq = list(inputs.values())[0]
target_length = len(target_seq)

binder_items = []
for name, seq in inputs.items():
    if name != 'target':
        binder_items.append((name, seq))

print(f"Target length: {target_length}")
print(f"Total binders: {len(binder_items)}")

# Find the longest binder sequence to use as reference
longest_binder = max(binder_items, key=lambda x: len(x[1]))
ref_name, ref_seq = longest_binder
ref_length = len(ref_seq)
print(f"Longest query (reference): {ref_name} (length: {ref_length})")
print("="*60)

# ----------------------------------------------------------------------------
# PART 2: Load target templates
# ----------------------------------------------------------------------------

with open(template_target_file, 'r') as f:
    target_templates = json.load(f)
print(f"Loaded {len(target_templates)} target templates")

# ----------------------------------------------------------------------------
# PART 3: ANARCI alignment functions
# ----------------------------------------------------------------------------

def collect_all_sequences(binder_items, paired_MSA_collection_path, unpaired_MSA_collection_path):
    """
    Collect ALL sequences from binders and the TWO MSA files.
    
    Args:
        binder_items: List of (name, seq) for all binders
        paired_MSA_collection_path: Path to the SINGLE paired MSA file (BINDER.a3m)
        unpaired_MSA_collection_path: Path to the SINGLE unpaired MSA file (BINDER.a3m)
    """
    all_sequences = []
    
    # Add all binder sequences
    for name, seq in binder_items:
        all_sequences.append(seq)
    
    # Read PAIRED MSA file
    with open(paired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            homo_seq = lines[i + 1].strip()
            if homo_seq:
                all_sequences.append(homo_seq)
    
    # Read UNPAIRED MSA file
    with open(unpaired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            homo_seq = lines[i + 1].strip()
            if homo_seq:
                all_sequences.append(homo_seq)
    
    return all_sequences


anarci_cache = {}

def get_anarci_numbering_dict(sequence, scheme='imgt'):
    """
    Arguments:
        sequence: Raw amino acid sequence (can have gaps - will be cleaned)
        scheme: Numbering scheme ('imgt', 'aho', etc.)
    
    Returns:
        Dictionary: {(pos, ins): aa, ...} or None if fails
        Example: {(1, ''): 'Q', (2, ''): 'V', (111, 'A'): 'D', ...}
    """
    # Clean sequence: remove gaps, keep only letters
    seq_clean = ''.join([c for c in sequence if c.isalpha()])
    
    # Skip sequences that are too short (not full antibodies)
    if len(seq_clean) < 50:
        return None
    
    # Check cache first - if we've seen this sequence before, return stored result
    if seq_clean in anarci_cache:
        return anarci_cache[seq_clean]
    
    # Run ANARCI (this is the slow part - happens only once per unique sequence)
    result = anarci([('seq', seq_clean)], scheme=scheme, output=False, allowed_species=[])

    if result is None:
        return None

    if isinstance(result, tuple) and len(result) >= 1:
        numbered_seqs = result[0]
      # print("Warning - multiple results per alignment found, only first used")
    else:
        numbered_seqs = result

    if not numbered_seqs or not numbered_seqs[0]:
        return None

    numbered = numbered_seqs[0][0][0]
    
    N = defaultdict(int)
    order = ' ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    corrected_numbered = []

    for (pos, ins), aa in numbered:
        if aa is not None:
            count = N[pos]
            correct_ins = order[count]
            corrected_numbered.append(((pos, correct_ins), aa))
            N[pos] += 1
    
    # Convert to dictionary for fast lookup
    pos_dict = {}
    for (pos, ins), aa in corrected_numbered:
        if aa is not None:
            pos_dict[(pos, ins)] = aa

    # Store in cache for future use
    anarci_cache[seq_clean] = pos_dict
    return pos_dict


def build_master_framework(all_sequences, scheme='imgt'):
    """
    Build a master framework containing ALL IMGT positions from all sequences.
    Uses cached ANARCI results to avoid redundant processing.
    
    Args:
        all_sequences: List of sequences (can include binders and homologs)
        scheme: Numbering scheme
    
    Returns:
        master_positions: Sorted list of all unique (pos, ins) tuples
    """
    all_positions = set()
    
    print(f"  Processing {len(all_sequences)} sequences to build master framework...")
    
    for seq in tqdm(all_sequences, desc="    Building master framework", unit="seq"):
        pos_dict = get_anarci_numbering_dict(seq, scheme)
        if pos_dict:
            for pos in pos_dict.keys():
                all_positions.add(pos)
                pos_num, pos_ins = pos
                if pos_ins != '' and pos_ins is not None:
                    all_positions.add((pos_num, ''))                     
    
    # Sort positions
    def sort_key(item):
        pos, ins = item
        ins_key = '' if ins is None or ins == '' else ins
        return (pos, ins_key)
    
    sorted_positions = sorted(all_positions, key=sort_key)
    
    # Clean: remove bare numbers that have lettered versions
    numbers_with_letters = set()
    for pos, ins in sorted_positions:
        if ins != '' and ins is not None:
            numbers_with_letters.add(pos)
    
    master_positions = []
    for pos, ins in sorted_positions:
        if ins == '' or ins is None:
            # Bare number - only keep if no lettered versions exist
            if pos not in numbers_with_letters:
                master_positions.append((pos, ins))
        else:
            # Lettered version - always keep
            master_positions.append((pos, ins))
    
    return master_positions

def align_to_master_framework(master_positions, pos_dict):
    """
    Arguments:
        master_positions: Sorted list of (pos, ins) tuples
        pos_dict: Dictionary mapping (pos, ins) to amino acid
    
    Returns:
        aligned_string: Sequence with residues at master positions, '-' elsewhere
        residue_count: Number of actual residues (for verification)
    """
    if not pos_dict:
        return None, 0
    
    # Build aligned sequence
    aligned_chars = []
    for position in master_positions:
        if position in pos_dict:
            aligned_chars.append(pos_dict[position])
        else:
            aligned_chars.append('-')
    
    aligned = ''.join(aligned_chars)
    residue_count = sum(1 for c in aligned if c.isalpha())
    
    return aligned, residue_count


def precompute_aligned_sequences(master_positions, binder_items, paired_MSA_collection_path, unpaired_MSA_collection_path, scheme='imgt'):
    """
    Precompute aligned versions of ALL sequences.
    
    Reads the SINGLE paired and SINGLE unpaired MSA files.
    """
    aligned_queries = {}
    precomputed_paired = []
    precomputed_unpaired = []
    paired_headers = []
    unpaired_headers = []
    paired_query_header = None
    unpaired_query_header = None
    
    # Track seen sequences to avoid duplicates
    seen_paired = set()
    seen_unpaired = set()
    
    # Store dictionaries for fast alignment later
    numbered_queries = {}
    
    print("  Getting numbering for binder queries...")
    for name, seq in tqdm(binder_items, desc="    Numbering binders", unit="binder"):
        pos_dict = get_anarci_numbering_dict(seq, scheme)
        if pos_dict:
            numbered_queries[seq] = pos_dict
    
    print("  Aligning binder queries to master framework...")
    for seq, pos_dict in numbered_queries.items():
        aligned, _ = align_to_master_framework(master_positions, pos_dict)
        if aligned:
            aligned_queries[seq] = aligned
    
    print("  Processing PAIRED MSA file...")
    with open(paired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    
    paired_query_header = lines[0]
    
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            header = lines[i]
            homo_seq = lines[i + 1].strip()
            
            paired_headers.append(header)
            
            raw_seq = ''.join([c for c in homo_seq if c.isalpha()])
            
            if raw_seq and raw_seq not in seen_paired:
                pos_dict = get_anarci_numbering_dict(raw_seq, scheme)
                if pos_dict:
                    aligned, _ = align_to_master_framework(master_positions, pos_dict)
                    if aligned:
                        precomputed_paired.append(aligned)
                        seen_paired.add(raw_seq)
    
    print("  Processing UNPAIRED MSA file...")
    with open(unpaired_MSA_collection_path, 'r') as f:
        lines = f.readlines()
    
    unpaired_query_header = lines[0]
    
    for i in range(2, len(lines), 2):
        if i + 1 < len(lines):
            header = lines[i]
            homo_seq = lines[i + 1].strip()
            
            unpaired_headers.append(header)
            
            raw_seq = ''.join([c for c in homo_seq if c.isalpha()])
            
            if raw_seq and raw_seq not in seen_unpaired:
                pos_dict = get_anarci_numbering_dict(raw_seq, scheme)
                if pos_dict:
                    aligned, _ = align_to_master_framework(master_positions, pos_dict)
                    if aligned:
                        precomputed_unpaired.append(aligned)
                        seen_unpaired.add(raw_seq)
    
    print(f"  Precomputed {len(precomputed_paired)} unique PAIRED homologs")
    print(f"  Precomputed {len(precomputed_unpaired)} unique UNPAIRED homologs")
    
    return (aligned_queries, precomputed_paired, precomputed_unpaired, 
            paired_headers, unpaired_headers, paired_query_header, unpaired_query_header)


def process_binder(binder_seq, master_positions, aligned_queries, precomputed_homologs, headers, query_header, scheme='imgt'):
    """
    Process a single binder: apply transformations and return MSA components.
    
    Arguments:
        binder_seq: Raw binder sequence
        master_positions: Master framework positions (for fallback alignment)
        aligned_queries: Dict mapping binder_seq to aligned version
        precomputed_homologs: List of precomputed aligned homologs
        headers: List of headers for homologs
        query_header: Header line for the query
        scheme: Numbering scheme
    
    Returns:
        raw_query: Raw query (for MSA line 2)
        processed_homologs: List of processed homologs ready for MSA
        msa_headers: List of headers for MSA
        msa_sequences: List of sequences for MSA
    """
    raw_query = binder_seq.upper()
    
    # Get aligned query from precomputed map
    query_aligned = aligned_queries[binder_seq]
    
    # Apply dummy lowercase and substitution
    processed_homologs = lowercase_substitute(query_aligned, precomputed_homologs)
    
    # Build MSA components
    msa_headers = [query_header]
    msa_sequences = [raw_query]
    
    for i in range(min(len(headers), len(processed_homologs))):
        msa_headers.append(headers[i])
        msa_sequences.append(processed_homologs[i])
    
    return raw_query, processed_homologs, msa_headers, msa_sequences


def lowercase_substitute(aligned_query, aligned_homologs):
    """
    Convert uppercase to lowercase in homolog where query has a dash
    """
    processed = []
    aligned_query = aligned_query.upper()
    
    for homo in aligned_homologs:
        temp_homo = []
        homo = homo.upper()
        
        for i in range(len(aligned_query)):
            q_char = aligned_query[i]
            h_char = homo[i]
            
            if q_char == '-' and h_char == '-':
                pass
            elif q_char == '-' and h_char.isalpha():
                temp_homo.append(h_char.lower())
            else:
                temp_homo.append(h_char)
        
        processed.append(''.join(temp_homo))
    
    return processed


def filter_msa_by_query_length(header_list, sequence_list, query_sequence):
    
    query_length = len(query_sequence)
    
    filtered_headers = [header_list[0]]
    filtered_sequences = [sequence_list[0]]
    removed = 0
    total = len(header_list) - 1  # Exclude query
    
    for i in range(1, len(header_list)):
        header = header_list[i]
        sequence = sequence_list[i]
        
        clean_sequence_chars = []
        for c in sequence:
            if not c.islower() or c == '-':
                clean_sequence_chars.append(c)
        clean_sequence = ''.join(clean_sequence_chars)
        
        if len(clean_sequence) == query_length:
            filtered_headers.append(header)
            filtered_sequences.append(sequence)
        else:
            removed += 1
    
    return filtered_headers, filtered_sequences, removed, total


def remove_gaps(sequence):
    
    result = []
    for c in sequence:
        if c.isalpha():
            result.append(c)
    return ''.join(result)


def check_indices(templates, seq_length):
    """Check that all indices in templates are within bounds."""
    all_valid = True
    for template in templates:
        if 'queryIndices' in template and template['queryIndices']:
            for i in template['queryIndices']:
                if i >= seq_length:
                    all_valid = False
                    break
        if not all_valid:
            break
            
    return all_valid


def get_safe_name(name):
    """Convert a binder name to a filesystem-safe string."""
    safe = ''
    for c in name:
        if c == '/' or c == '\\' or c == '.':
            safe += '_'
        else:
            safe += c
    return safe


def get_paired_path(safe_name):
    """Return the path to the paired MSA file for a binder."""
    return f"{MSA_collection_path}/pairedMsa/binder/{safe_name}.a3m"


def get_unpaired_path(safe_name):
    """Return the path to the unpaired MSA file for a binder."""
    return f"{MSA_collection_path}/unpairedMsa/binder/{safe_name}.a3m"

# ----------------------------------------------------------------------------
# PART 4: BUILD MASTER FRAMEWORK
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 1: Building master framework from all sequences")
print("="*60)

start_time = time.time()

all_sequences = collect_all_sequences(binder_items, pairedMsa_binder_file, unpairedMsa_binder_file)
print(f"Collected {len(all_sequences)} total sequences")

master_positions = build_master_framework(all_sequences)
print(f"Master framework has {len(master_positions)} positions")
print(f"Master framework build time: {time.time() - start_time:.1f}s")

# ----------------------------------------------------------------------------
# PART 5: PRECOMPUTE ALIGNED SEQUENCES
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 2: Precomputing aligned sequences")
print("="*60)

start_time = time.time()

(aligned_queries, precomputed_paired, precomputed_unpaired, 
 paired_headers, unpaired_headers, paired_query_header, unpaired_query_header) = precompute_aligned_sequences(
    master_positions, binder_items, pairedMsa_binder_file, unpairedMsa_binder_file)

print(f"Precomputation time: {time.time() - start_time:.1f}s")

# ----------------------------------------------------------------------------
# PART 6: Process each binder using trimmed alignments
# ----------------------------------------------------------------------------

print("\n" + "="*60)
print("PHASE 3: Processing individual binders")
print("="*60)

total_start_time = time.time()

for binder_idx, (original_name, binder_seq) in enumerate(tqdm(binder_items, desc="Processing binders", unit="binder"), 1):
    binder_start_time = time.time()
    
    # Use helper function for safe name
    safe_name = get_safe_name(original_name)
    
    binder_length = len(binder_seq)
    raw_query = binder_seq.upper()
    
    print(f"\n[{binder_idx}/{len(binder_items)}] Processing {original_name} (length: {binder_length})...")
    
    # --- Create binder-specific template file with trimmed indices ---
    with open(template_binder_file, 'r') as f:
        binder_templates = json.load(f)
    
    for template in binder_templates:
        if 'queryIndices' in template:
            trimmed_indices = []
            for i in template['queryIndices']:
                if i < binder_length:
                    trimmed_indices.append(i)
            template['queryIndices'] = trimmed_indices
    
    binder_template_copy = f"{MSA_collection_path}/templates/binder/{safe_name}_templates.json"
    os.makedirs(os.path.dirname(binder_template_copy), exist_ok=True)
    
    with open(binder_template_copy, 'w') as f:
        json.dump(binder_templates, f, indent=2)
    
    # --- Process PAIRED MSA ---
    paired_start_time = time.time()
    print(f"  [PAIRED MSA] Processing...")
    
    raw_query, processed_homologs, msa_headers, msa_sequences = process_binder(
        binder_seq, master_positions, aligned_queries, precomputed_paired, 
        paired_headers, paired_query_header
    )
    
    filtered_headers, filtered_sequences, paired_removed, paired_total = filter_msa_by_query_length(
        msa_headers, msa_sequences, raw_query
    )
    
    paired_file = get_paired_path(safe_name)
    os.makedirs(os.path.dirname(paired_file), exist_ok=True)
    
    with open(paired_file, 'w') as f:
        for j in range(len(filtered_headers)):
            f.write(filtered_headers[j])
            f.write(filtered_sequences[j] + '\n')
    
    paired_passed = len(filtered_sequences) - 1
    paired_time = time.time() - paired_start_time
    print(f"  [PAIRED MSA] ✓ Passed: {paired_passed}/{paired_total} | Removed: {paired_removed} | Time: {paired_time:.1f}s")
    
    # --- Process UNPAIRED MSA ---
    unpaired_start_time = time.time()
    print(f"  [UNPAIRED MSA] Processing...")
    
    raw_query, processed_homologs, msa_headers, msa_sequences = process_binder(
        binder_seq, master_positions, aligned_queries, precomputed_unpaired, 
        unpaired_headers, unpaired_query_header
    )
    
    filtered_headers, filtered_sequences, unpaired_removed, unpaired_total = filter_msa_by_query_length(
        msa_headers, msa_sequences, raw_query
    )
    
    unpaired_file = get_unpaired_path(safe_name)
    os.makedirs(os.path.dirname(unpaired_file), exist_ok=True)
    
    with open(unpaired_file, 'w') as f:
        for j in range(len(filtered_headers)):
            f.write(filtered_headers[j])
            f.write(filtered_sequences[j] + '\n')
    
    unpaired_passed = len(filtered_sequences) - 1
    unpaired_time = time.time() - unpaired_start_time
    print(f"  [UNPAIRED MSA] ✓ Passed: {unpaired_passed}/{unpaired_total} | Removed: {unpaired_removed} | Time: {unpaired_time:.1f}s")
    
    # --- Create JSON files ---
    with open(binder_template_copy, 'r') as f:
        trimmed_binder_templates = json.load(f)

    query_no_gaps = remove_gaps(aligned_queries.get(binder_seq, binder_seq))

    json_parts = []
    json_parts.append('{\n')
    json_parts.append(f'  "name": "{original_name}",\n')
    json_parts.append('  "modelSeeds": [1, 2],\n')
    json_parts.append('  "sequences": [\n')
    json_parts.append('    {\n')
    json_parts.append('      "protein": {\n')
    json_parts.append('        "id": "TARGET",\n')
    json_parts.append(f'        "sequence": {json.dumps(target_seq)},\n')
    json_parts.append(f'        "unpairedMsaPath": {json.dumps(unpairedMsa_target_file)},\n')
    json_parts.append(f'        "pairedMsaPath": {json.dumps(pairedMsa_target_file)},\n')
    json_parts.append('        "templates": [\n')
    
    for i in range(len(target_templates)):
        template = target_templates[i]
        json_parts.append('  {\n')
        json_parts.append(f'    "mmcifPath": {json.dumps(template["mmcifPath"])},\n')
        
        query_indices_str = '['
        for j in range(len(template['queryIndices'])):
            if j > 0:
                query_indices_str += ', '
            query_indices_str += str(template['queryIndices'][j])
        query_indices_str += ']'
        json_parts.append(f'    "queryIndices": {query_indices_str},\n')
        
        template_indices_str = '['
        for j in range(len(template['templateIndices'])):
            if j > 0:
                template_indices_str += ', '
            template_indices_str += str(template['templateIndices'][j])
        template_indices_str += ']'
        json_parts.append(f'    "templateIndices": {template_indices_str}\n')
        
        if i < len(target_templates) - 1:
            json_parts.append('  },\n')
        else:
            json_parts.append('  }\n')
    
    json_parts.append(']\n')
    json_parts.append('      }\n')
    json_parts.append('    },\n')
    json_parts.append('    {\n')
    json_parts.append('      "protein": {\n')
    json_parts.append('        "id": "BINDER",\n')
    json_parts.append(f'        "sequence": {json.dumps(query_no_gaps)},\n')
    json_parts.append(f'        "unpairedMsaPath": {json.dumps(unpaired_file)},\n')
    json_parts.append(f'        "pairedMsaPath": {json.dumps(paired_file)},\n')
    json_parts.append('        "templates": [\n')
    
    for i in range(len(trimmed_binder_templates)):
        template = trimmed_binder_templates[i]
        json_parts.append('  {\n')
        json_parts.append(f'    "mmcifPath": {json.dumps(template["mmcifPath"])},\n')
        
        query_indices_str = '['
        for j in range(len(template['queryIndices'])):
            if j > 0:
                query_indices_str += ', '
            query_indices_str += str(template['queryIndices'][j])
        query_indices_str += ']'
        json_parts.append(f'    "queryIndices": {query_indices_str},\n')
        
        template_indices_str = '['
        for j in range(len(template['templateIndices'])):
            if j > 0:
                template_indices_str += ', '
            template_indices_str += str(template['templateIndices'][j])
        template_indices_str += ']'
        json_parts.append(f'    "templateIndices": {template_indices_str}\n')
        
        if i < len(trimmed_binder_templates) - 1:
            json_parts.append('  },\n')
        else:
            json_parts.append('  }\n')
    
    json_parts.append(']\n')
    json_parts.append('      }\n')
    json_parts.append('    }\n')
    json_parts.append('  ],\n')
    json_parts.append('  "dialect": "alphafold3",\n')
    json_parts.append('  "version": 4\n')
    json_parts.append('}')
    
    json_file = os.path.join(json_collection_path, f"{safe_name}.json")
    dir_path = ''
    for part in json_file.split('/')[:-1]:
        dir_path += part + '/'
    if dir_path:
        os.makedirs(dir_path, exist_ok=True)
    
    with open(json_file, 'w') as f:
        f.write(''.join(json_parts))
    
    binder_time = time.time() - binder_start_time
    print(f"  ✓ JSON created: {safe_name}.json | Binder time: {binder_time:.1f}s")
    print("-" * 60)
    
# ----------------------------------------------------------------------------
# FINAL SUMMARY
# ----------------------------------------------------------------------------

total_time = time.time() - total_start_time
print("\n" + "="*60)
print("PIPELINE COMPLETE")
print("="*60)
print(f"Total processing time: {total_time/60:.1f} minutes ({total_time:.1f} seconds)")
print(f"Output files:")
print(f"  Paired MSAs: {MSA_collection_path}/pairedMsa/binder/")
print(f"  Unpaired MSAs: {MSA_collection_path}/unpairedMsa/binder/")
print(f"  JSON files: {json_collection_path}")
print("="*60)
```

```python
hmmer_path = "/cluster/tufts/cowenlab/jreill05/Database_Conda_Environment/condaenv/AlphaFold3/bin"
os.environ["PATH"] = hmmer_path + ":" + os.environ.get("PATH", "")
from anarci import anarci

import time
from tqdm import tqdm
from collections import defaultdict

anarci_cache = {}

def get_anarci_numbering_dict(sequence, scheme='imgt'):
    # Clean sequence
    seq_clean = ''.join([c for c in sequence if c.isalpha()])
    
    if len(seq_clean) < 50:
        return None
    
    # Check cache
    if seq_clean in anarci_cache:
        return anarci_cache[seq_clean]
    
    # Run ANARCI
    result = anarci([('seq', seq_clean)], scheme=scheme, output=False, allowed_species=[])
    
    if result is None:
        return None
    
    if isinstance(result, tuple) and len(result) >= 1:
        numbered_seqs = result[0]
    else:
        numbered_seqs = result
    
    if not numbered_seqs or not numbered_seqs[0]:
        return None
    
    numbered = numbered_seqs[0][0][0]
    
    N = defaultdict(int)
    order = ' ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    corrected_numbered = []

    for (pos, ins), aa in numbered:
        if aa is not None:
            count = N[pos]
            correct_ins = order[count]
            corrected_numbered.append(((pos, correct_ins), aa))
            N[pos] += 1
    
    # Convert to dictionary for fast lookup
    pos_dict = {}
    for (pos, ins), aa in corrected_numbered:
        if aa is not None:
            pos_dict[(pos, ins)] = aa

    anarci_cache[seq_clean] = pos_dict
    return pos_dict

query1 = 'SVQAGGSLRLSCAASGTFFRINYMGWYRQTPENQQEMVAIIDSEMRADYSDSVQGRFAISRDNAKNTVYLEMSSLKPEDTAIYFCNAKAVTTSPQYRMADYWGQGTRVTVSS'

get_anarci_numbering_dict(query1)
```
